```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    step = 1

    print(f"\nШукаємо число {target} у відсортованому масиві: {arr}")

    while low <= high:
        mid = (low + high) // 2
        guess = arr[mid]

        print(f"Крок {step}: перевіряємо індекс {mid} (значення {guess})")

        if guess == target:
            print(f"--- Результат: Число {target} знайдено на позиції {mid}! ---")
            return mid
        if guess > target:
            high = mid - 1
        else:
            low = mid + 1
        step += 1

    print(f"--- Результат: Числа {target} немає в списку. ---")
    return None

# Введення даних
try:
    input_str = input("Введіть числа через кому (наприклад: -5, 9, 0, -32, 45, 39): ")
    # Перетворюємо рядок у список цілих чисел та сортуємо його
    data = sorted([int(x.strip()) for x in input_str.split(',')])

    target_num = int(input("Введіть число для пошуку: "))

    binary_search(data, target_num)
except ValueError:
    print("Помилка: будь ласка, вводьте тільки числа через кому.")
```
